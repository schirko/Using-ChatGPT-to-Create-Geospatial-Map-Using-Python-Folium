## Getting Started

First we need to import pandas to get the data and folium to help create the map. In this case, we asked ChatGPT to filter the data so that we only received airports in Colorado and those that were not classified as small.

```python
import folium
import pandas as pd

# Read CSV file
data = pd.read_csv("us-airports.csv")

# Filter the data to only include locations in Colorado
data = data[(data['region_name'] == 'Colorado') & (data['type'] != 'small_airport')]
```

# Create the map

To create the map we first need the longitude and latitude from the dataset.  Next ChatGPT uses a for loop to gather all the records in the dataset.

```python
# Create map
map = folium.Map(location=[data['latitude_deg'].mean(), data['longitude_deg'].mean()], zoom_start=7)

# Add markers for each data point
for i in range(0, len (data)):
    folium.Marker([data.iloc[i]['latitude_deg'], data.iloc[i]['longitude_deg']],popup=data.iloc[i]['name']).add_to(map)
```

# Save the map to HTML and display it

```python
# Save map to HTML file
map.save("colorado-airports.html")    
    
# Display the map
map
```
